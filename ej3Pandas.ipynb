```python
# magic function para hacer que los graficos de matplotlib se renderizen en el notebook.
%matplotlib notebook

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
person_data = {
        'codigo_producto': ['95001', '95002', '95003', '95004', '95005','95005','95007','95007','95008','95009'],
        'codigo_fabricante': ['F95001', 'F95001', 'F95003', 'F95003', 'F95001','F95005','F95003','F95009','F95009','F95009'],
        'mes': ['08', '04', '08', '05', '11','04','12','10','10','02'],
        'ventas_mensuales':['1200','850','345','200','3000','27000','45000','900','1500','1']}

df_a = pd.DataFrame(person_data, columns = ['codigo_producto', 'codigo_fabricante','mes','ventas_mensuales'])
```

```python
#Paso a enteros la columna ventas mensuales para poder calcular promedios
df_a["ventas_mensuales"]=df_a["ventas_mensuales"].astype(int)
df_a
```

```python
#Agrupo por codigo de producto y calculo el promedio de ventas por producto (punto 3A)
groupPromedio_Ventas=df_a.groupby(["codigo_producto"]).agg({"ventas_mensuales":"mean"})
groupPromedio_Ventas.rename(columns={"ventas_mensuales":"promedio_ventas"})
```

```python
#Ahora agrupo por producto y calculo el minimo el maximo y promedio de las ventas mensuales 
groupPromedio_Ventas=df_a.groupby(["codigo_producto"])["ventas_mensuales"].agg(["min","max","mean"])
df_a2=groupPromedio_Ventas.rename(columns={"min":"minimo_ventas_mensual","max":"maximo_ventas_mensual","mean":"promedio_ventas_mensual"})
df_a2
```

```python
#Genero la tabla pedida (punto 3C)
df_a2.reset_index()
```

```python
#Genero la tabla pedida (punto 3D)
df_a2.T
```

```python
#Genero la tabla pedida (punto 3B)
df_b=df_a2.stack()
df_c=df_b.to_frame().rename(columns={0:"montos"})
df_c
```
