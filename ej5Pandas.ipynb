```python
# magic function para hacer que los graficos de matplotlib se renderizen en el notebook.
%matplotlib notebook

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
hospitales_data = {
        'dia_nacimiento': ['01', '02', '03', '04', '05','05','07','07','08','09','10'],
        'mes_nacimiento': ['10', '10', '10', '10', '10','05','10','09','09','10','10'],
        'anio_nacimiento': ['2017', '2017', '2016', '2016', '2017','2017','2017','2017','2017','2018','2017'],
        'peso_al_nacer': ['1.5', '2', '3', '4', '2.5','2.5','1.2','1.10','1.9','2.9','3'],
        'longitud_al_nacer': ['25', '27', '35', '36', '25','25','29','34','39','29','35'],
        'id_hospital': ['H350', 'H250', 'H303', 'H304', 'H250','H205','H350','H250','H205','H303','H304'],
        'tipo_parto': ['1', '2', '2', '1', '2','2','1','1','1','2','2'],

}

df_a = pd.DataFrame(hospitales_data, columns = ['dia_nacimiento', 'mes_nacimiento','anio_nacimiento','peso_al_nacer','longitud_al_nacer','id_hospital','tipo_parto'])

hospitales_historia = {
        'id_hospital': ['H350', 'H250', 'H303', 'H304', 'H205','H205','H350','H250','H205','H303'],
        'direccion' : ['Espindola 900','Saraza 456','Malvinas 3550','Diamante 1570','Mengano 789','Perinola 123','Espindola 900','Saraza 456','Sacachispa 250','Queseyo 25'],
        'promedio_nacimientos_mensual' : ['1','3','4','6','7','4','1','3','23','2'],}

df_b = pd.DataFrame(hospitales_historia, columns = ['id_hospital', 'direccion','promedio_nacimientos_mensual'])
```

```python
df_b
```

```python
df_a2=df_a.loc[df_a["mes_nacimiento"]=='10']
df_a3=df_a2.loc[df_a2["anio_nacimiento"]=='2017']
df_a3
```

```python
df_a4=df_a3.groupby(['id_hospital']).agg({'id_hospital':'count'})
df_a4.rename(columns={"id_hospital":"cantidad_nacimientos"},inplace=True)
df_a5=df_a4.reset_index()
df_a5
```

```python
df_m3=pd.merge(df_a5, df_b, on='id_hospital', how='left')
df_m3
```

```python
df_m3.info()
```

```python
df_m3['promedio_nacimientos_mensual']=df_m3.loc[:,'promedio_nacimientos_mensual'].astype(int)
df_m3
```

```python
df_m3.info()
```

```python
df_m4=df_m3.loc[df_m3["cantidad_nacimientos"]>df_m3["promedio_nacimientos_mensual"]]
df_m4
```

```python
df_m4.drop_duplicates()
```
