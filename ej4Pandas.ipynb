```python
# magic function para hacer que los graficos de matplotlib se renderizen en el notebook.
%matplotlib notebook

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
ebooks_data = {
        'user_id': ['95001', '95002', '95003', '95004', '95005','95005','95007','95007','95008','95009'],
        'book_id': ['F95001', 'F95001', 'F95003', 'F95003', 'F95001','F95005','F95003','F95009','F95009','F95009'],
        'rating': ['08', '04', '08', '05', '10','04','09','10','10','02'],
        'timestamp': ['01-08-2017', '02-04-2016', '30-08-2019', '04-05-2018', '25-11-2019','25-04-2019','12-12-2019','10-10-2019','19-10-2019','29-02-2019'],}

df_a = pd.DataFrame(ebooks_data, columns = ['user_id', 'book_id','rating','timestamp'])

Goodreads = {
        'book_id': ['F95001', 'F95001', 'F95003', 'F95003', 'F95001','F95005','F95003','F95009','F95009','F95009'],
        'book_name': [ 'El Principito','El señor de los anillos', 'Niesztche Biografia', 'Cuentos de Galeano', 'Las aventuras de Tom Sawyer','Isaac Asimov','Fisica 1','Matematica 1','Roberto Fontana','Manual Linux'],
        'avg_rating': ['05', '07', '04', '10', '05','08','03','10','07','08'],}

df_b = pd.DataFrame(Goodreads, columns = ['book_id', 'book_name','avg_rating'])
```

```python
df_a
```

```python
df_a.info()
```

```python
df_a2=df_a.loc[:,'rating'].astype('int')
df_a['rating']=df_a2
df_a
```

```python
df_a3=df_a.groupby(["book_id"]).agg({"rating":"mean"})
df_a4=df_a3.reset_index()
df_a4
```

```python
df_b
```

```python
df_b2=df_b.loc[:,'avg_rating'].astype(int)
df_b['avg_rating']=df_b2
df_b
```

```python
df_b3=df_b.groupby(['book_id']).agg({"avg_rating":"mean"})
df_b4=df_b3.reset_index()
df_b4
```

```python
df_m=pd.merge(df_a4, df_b4, on='book_id', how='inner')
df_m
```

```python
df_m["promedio>20"]=((df_m["rating"]-df_m["avg_rating"]).abs())>2
df_m
```

```python
df_m2=df_m.loc[df_m["promedio>20"]==True,:]
df_m2
```

```python
df_m3=pd.merge(df_b, df_m2, on='book_id', how='right')
df_m3
```
